## PART 1

Plotting and describing the shape of the salary data.

```python
# import the data
data = pd.read_csv('ds_salary.csv')
data.hist(column='salary_in_usd', bins=20)
print("Median Salary in USD: ", data["salary_in_usd"].median())
print("Mean Salary in USD: ", data["salary_in_usd"].mean())
print("Highest Salary in USD: ", data["salary_in_usd"].max())
print("Lowest Salary in USD: ", data["salary_in_usd"].min())
```

**Analysis of salary data**:

Based on the raw USD salary data (without relation to any other data features), we can see that the majority of data scientists receive a compensation close to 100,000 USD. The median salary is 105,000 USD with the mean being slightly higher at ~116,000 USD.

The higher mean salary is likely due to a few outliers who are compensated above 300,000 USD. However, we can see that a salary this high is unusual and therefore, the median is probably a better indication of Data Scientist salary.

Finally we have our highest salary coming in at 600,000 USD and our lowest salary coming in at 2,859 USD. I think the lowest salary must be a mistake because that is only ~55USD/wk if it is full time annual salary.

## Part 2

**Effect of experience on salary**

I will be using column "salary_in_usd" for salary and "experience_level" for experience

```python
# create the scatterplot of the salary in USD vs. experience
data.plot.scatter(x='experience_level', y='salary_in_usd')

# max/min/mean/median salary for each experience level
print("Maximum Salary per experience level: ", data.groupby('experience_level')['salary_in_usd'].max())
print("Minimum Salary per experience level: ", data.groupby('experience_level')['salary_in_usd'].min())
print("Average Salary per experience level: ", data.groupby('experience_level')['salary_in_usd'].mean())
print("Median Salary per experience level: ", data.groupby('experience_level')['salary_in_usd'].median())
```

### Analysis

The discrete data experience variables make it a little more difficult to easily see trends in the scatterplot. However, by also including the respective means, medians, maxes, and mins, we are able to see a more obvious trend.

**Define The Experience Variables**
- EN Entry-level or Junior
- MI Mid-level or Intermediate
- SE Senior-level or Expert
- EX Executive-level or Director

#### **Describing The Visuals**

##### EN or Entry-Level
The scatterplot clearly shows that the EN or entry-level engineers have a salary that is clustered down near the 75,000 range. There are also very few high outliers, suggesting a pretty firm salary ceiling for entry level data scientists. The lowest salary employee is either getting scammed or it is a mistake in the data. He is making less than 2% of the top earner's salary, and ~7% of the avg salary.

- Max: 250,000 USD
- Min: 4,000 USD
- Avg: ~65,250 USD
- Median: 59,551 USD

##### MI or Mid-Level
The plot shows that the majority of data scientists at the mid-level earn a larger salary than the entry-level scientists (avg increase of ~39%), and there is a much higher salary ceiling (this is likely due to outstanding engineers who are compensated as such).

Once again, the lowest salary guy must be a mistake because this salary is nearly half of the lowest entry level salary.

- Max: 450,000 USD
- Min: 2,859 USD
- Avg: ~91,100 USD
- Median: 78,526 USD

##### SE or Senior-Level
The plot shows a generally higher spread for senior data scientists than mid-level scientists. While the top paid person was compensated roughly the same, the average shows a notable increase in salary for senior vs mid-level.

- Max: 412,000 USD
- Min: 18,907 USD
- Avg: ~140,100 USD
- Median: 136620 USD

##### EX or Expert-Level/Director
There are far fewer data points for expert level or director level data scientists. However, these scientists are generally paid much more than the lower levels (Senior, Mid-Level, Entry-Level).

- Max: 600,000 USD
- Min: 69,741 USD
- Avg: ~204,200 USD
- Median: 175,000 USD

### OVERALL

Overall we can see that as experience level increases, the salary will also increase. We see that the largest increase in average salary is from mid-level to senior data scientist with over 50% avg salary increase, and the smallest increase in average salary is from entry-level to mid-level with only roughly 39% average salary increase.
